```python
import numpy as np

class SVM:
  def __init__(self,learning_rate,C,iters):
    self.learning_rate=learning_rate
    self.C=C
    self.iters=iters
    self.w=None
    self.b=None


  def bias(self,x):
     #adding bias
    one = np.ones((len(x),1))
    x = np.hstack((one,x))
    return x

  def gradient(self,x,y):
    #adding weights , changing labels, and gradient descent
    y=np.where(y<=0,-1,1)
    x = self.bias(x)
    self.w = np.zeros(x.shape[1])
    self.b=0
    #gradient descent (Stochastic Gradient Descent)
    for iters in range(self.iters):
      for i in range(len(y)):
        condition=y[i] * (np.dot(x[i], self.w)-self.b)
        if(condition>=1):
          dw=2*self.C*self.w
          db=0
        else:
          dw=2*self.C*self.w-np.dot(x[i],y[i])
          db=y[i]
        self.w=self.w-self.learning_rate*dw # Moved inside the inner loop
        self.b=self.b-self.learning_rate*db # Moved inside the inner loop

  def predict(self, x):
      x_bias = self.bias(x) # Add bias to the input for prediction
      output = np.dot(x_bias, self.w)-self.b
      predicted = np.sign(output)
      y_pred = np.where(predicted==-1,0,1)
      return y_pred

  def accuracy(self, X, y):
    preds = self.predict(X)
    return np.mean(preds == y)
```

```python
from sklearn.model_selection import train_test_split

x = df[['age', 'heart_disease', 'avg_glucose_level']].values
y = df['stroke'].values

# Original splitting method (commented out as requested):
# ratio=0.8
# split = int(ratio * len(x))
# x_train, x_test = x[:split], x[split:]
# y_train, y_test = y[:split], y[split:]

# Using sklearn's train_test_split with stratification
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)
```

```python
classifier = SVM(learning_rate=0.001,C=0.01,iters=1000)
classifier.gradient(x_train,y_train)
```

```python
pred = classifier.predict(x_test)
acc = classifier.accuracy(x_test, y_test)
print(acc)
```
